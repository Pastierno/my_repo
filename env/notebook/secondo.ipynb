```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Caricamento dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Standardizzazione delle features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Suddivisione del dataset in training e test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Applico algoritmo Decision Tree
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# Valutiamo performance del modello
y_pred = clf.predict(X_test)
report = classification_report(y_test, y_pred, target_names=iris.target_names)
print(report)
# accuracy: Quota predizioni corrette
# Precision Fra le predizioni positive, quante sono corrette?
# Recall: Fra tutti i casi positivi, quali sono stati predetti correttamente?
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
